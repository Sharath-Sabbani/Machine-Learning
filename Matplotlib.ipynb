```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
plt.scatter(x,y)
plt.plot(x,y,'r')
plt.show()
```

```python
x1=[1,2,3,4,5,6]
x2=[1,2,3,4,5,6]
y1=[1,2,3,4,5,6]
y2=[6,5,4,3,2,1]
plt.scatter(x1,y1)
plt.plot(x1,y1,"r*")  #(x1,y1,"r--")
plt.scatter(x2,y2)
plt.plot(x2,y2,"k+")  #(x2,y2,"r-.")
plt.show()
```

```python
x=np.array([1,2,3,4,5,6])
y=2*(x**2)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=2*(x**2)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.1)
y1=2*(x**2)
y2=3*(x**3)
plt.plot(x,y1,color="red",label="x^2",marker="*",linewidth=1)
plt.plot(x,y2,color="yellow",label="x^3",marker="+",linewidth=0.75)
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Prepinsta chart")
plt.axis([0,15,0,5000])
plt.grid()
plt.text(2,1000,"Hello Colab",fontsize=10)
plt.show()
```

# PIE CHARTS:

```python
profit = [10,20,30,25]
company = ["Flipkart","Amazon","Snapdeal","Jiomart"]
colors=["red","green","blue","yellow"]
explode=[0.05,0.1,0,0]
plt.pie(profit,colors=colors,labels=company,autopct="%.2f%%",counterclock=True,explode=explode,startangle=90)
plt.show()
```
