# Multiple Linear Regression

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Datasets

```python
data_frame = pd.read_csv("Covid_data.csv")
```

```python
X = data_frame.iloc[:,:4].values
y = data_frame.iloc[:,4].values
```

```python
X
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[1])],remainder="passthrough")
X = ct.fit_transform(X)
```

## Splitting in training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=7)
```

## Training the model using Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## Predicting the test values

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=3)
a = y_pred.reshape(len(y_pred),1)
b = y_test.reshape(len(y_test),1)
print(np.concatenate((a,b),1))
```

## Visualize data

```python
plt.plot(y_train)
plt.plot(regressor.predict(X_train),"g")
plt.show()
```

```python
plt.plot(y_test,color="red")
plt.plot(y_pred,color="blue")
plt.show()
```
