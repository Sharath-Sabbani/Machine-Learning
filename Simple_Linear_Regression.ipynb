# Linear Regression

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Datasets

```python
from pandas._libs.tslibs.offsets import YearBegin
data_set = pd.read_csv("Salar.csv")
X = data_set.iloc[:,0:1].values
y = data_set.iloc[:,1:2].values
```

## Splitting data in test and training set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Training the training set using Simple Linear regreesion

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## Predict the test data

```python
y_pred = regressor.predict(X_test)
```

## Plot the trained data

```python
plt.scatter(X_train,y_train,color="red")
plt.plot(X_train,regressor.predict(X_train),color="blue")
plt.show()
```

## Plot the test data

```python
plt.scatter(X_test,y_test,color="red")
plt.plot(X_test,y_pred,color="blue")
plt.show()
```
